```python
df_dataset = pd.read_csv('D:/Codes/tcc-sentiment-analysis/data/processed/dataset.csv')
```

```python
def separar_dataframe(df, proporcao_treino=0.8, random_state=None):
    """
    Separa o DataFrame em conjuntos de treino e teste.

    Args:
        df (pd.DataFrame): DataFrame a ser dividido.
        proporcao_treino (float): Proporção de dados de treino (entre 0 e 1).
        random_state (int, optional): Semente para reprodução da aleatoriedade.

    Returns:
        pd.DataFrame, pd.DataFrame: DataFrames de treino e teste.
    """
    # Verifica se a proporção está entre 0 e 1
    if not (0 < proporcao_treino < 1):
        raise ValueError("A proporção de treino deve estar entre 0 e 1.")
    
    # Separar os dados
    df_treino, df_teste = train_test_split(df, train_size=proporcao_treino, random_state=random_state)
    
    return df_treino, df_teste

df_treino, df_teste = separar_dataframe(df_dataset, proporcao_treino=0.75, random_state=42)
```
